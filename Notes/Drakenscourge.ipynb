Code Breakdown


Create a Character Class:


This defines how a character (like the hero or dragon) will work.
Each character has:
Name: What the character is called.
Health: How much life the character has.
Attack Power: How much damage the character can deal.

Create an Attack Function:


This function lets a character attack another character.
It calculates the damage and reduces the opponent's health.

Create a Heal Function:


This function allows a character to heal itself.
It randomly adds some health back to the character.

Game Function:


Set up the game:
Create a hero and a dragon, giving them names, health, and attack power.
Start a loop where each character takes turns.
Hero's Turn: The player chooses to attack or heal.
Dragon's Turn: The dragon randomly chooses to attack or heal.

End the Game:

If either the hero or the dragon's health reaches zero, the game ends.

```python
import random

class Character:
    def __init__(self, name, health, attack_power):
        self.name = name
        self.health = health
        self.attack_power = attack_power

    def attack(self, opponent):
        damage = self.attack_power
        print(f"{self.name} attacks {opponent.name} for {damage} damage.")
        opponent.health -= damage
        print(f"{opponent.name} has {opponent.health} health remaining.\n")

    def heal(self):
        heal_amount = random.randint(10, 30)
        self.health += heal_amount
        print(f"{self.name} heals for {heal_amount} health and now has {self.health} health.\n")

def game():
    print("Welcome to Drakenscourge!")

    # Creating the player and dragon
    player = Character("Hero", 100, 20)
    dragon = Character("Dragon", 100, 15)

    while player.health > 0 and dragon.health > 0:
        print("Choose your action: (1) Attack (2) Heal")
        choice = input("Enter your choice: ")

        if choice == "1":
            player.attack(dragon)
        elif choice == "2":
            player.heal()
        else:
            print("Invalid choice, please choose 1 or 2.\n")
            continue

        # Dragon's turn
        if dragon.health > 0:
            dragon_action = random.choice(["attack", "heal"])
            if dragon_action == "attack":
                dragon.attack(player)
            elif dragon_action == "heal":
                dragon.heal()

    # Game Over
    if player.health <= 0:
        print("You have been defeated by the Dragon.")
    else:
        print("You have defeated the Dragon! Victory is yours!")

# Start the game
game()
```
