Write a Python program to:

Add two numbers
Subtract two numbers
Multiply two numbers
Divide two numbers and return both the remainder and the quotient.

```python
def add_numbers(a, b):
    return a + b

def subtract_numbers(a, b):
    return a - b

def multiply_numbers(a, b):
    return a * b

def divide_numbers(a, b):
    quotient = a // b
    remainder = a % b
    return quotient, remainder

# Example usage:
num1 = 10
num2 = 3

print(f"Addition: {add_numbers(num1, num2)}")
print(f"Subtraction: {subtract_numbers(num1, num2)}")
print(f"Multiplication: {multiply_numbers(num1, num2)}")
quotient, remainder = divide_numbers(num1, num2)
print(f"Division: Quotient = {quotient}, Remainder = {remainder}")
```

Write a program to detect double spaces in a string.

```python
# Function to detect double spaces in a string
def detect_double_spaces(input_string):
    return "  " in input_string

# Get input from the user
user_input = input("Enter a string: ")

# Check for double spaces
has_double_spaces = detect_double_spaces(user_input)

# Display the result
if has_double_spaces:
    print("The string contains double spaces.")
else:
    print("The string does not contain double spaces.")
```

```python
# Example set containing a list
S = {(8, 7, 12), 'Harry', [1, 2]}

# Modify the list inside the set
for item in S:
    if isinstance(item, list):
        item[0] = 99  # Change the first element of the list

print(S)  # Output: {(8, 7, 12), 'Harry', [99, 2]}
```

Write a program to find the greatest of four numbers entered by the user.

```python
# Function to find the greatest of four numbers
def find_greatest(a, b, c, d):
    greatest = a
    if b > greatest:
        greatest = b
    if c > greatest:
        greatest = c
    if d > greatest:
        greatest = d
    return greatest

# Input four numbers from the user
num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))
num3 = float(input("Enter the third number: "))
num4 = float(input("Enter the fourth number: "))

# Find the greatest number
greatest_number = find_greatest(num1, num2, num3, num4)

# Output the result
print(f"The greatest number is: {greatest_number}")
```

Write a program to find out whether a student is pass or fail if it requires a total of 40% and at least 33% in each subject to pass. Assume three subjects and take marks as input from the user.

```python
# Function to check if the student has passed or failed
def check_pass_fail(marks):
    total_marks = sum(marks)
    percentage = (total_marks / 300) * 100

    # Check if the student has at least 33% in each subject
    for mark in marks:
        if mark < 33:
            return "Fail"

    # Check if the total percentage is at least 40%
    if percentage >= 40:
        return "Pass"
    else:
        return "Fail"

# Input marks for three subjects from the user
marks = []
for i in range(1, 4):
    mark = float(input(f"Enter marks for subject {i}: "))
    marks.append(mark)

# Determine if the student has passed or failed
result = check_pass_fail(marks)

# Output the result
print(f"The student has {result}.")
```

Write a program to find the factorial of a given number using a for loop.

```python
# Function to calculate factorial
def factorial(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

# Input a number from the user
number = int(input("Enter a number: "))

# Calculate the factorial
fact = factorial(number)

# Output the result
print(f"The factorial of {number} is {fact}")
```

Write a program to display A/B, where A and B are integers. If B is equal to 0, display "infinite" by handling the zero division error.

```python
def divide(A, B):
    try:
        result = A / B
        print(f"The result of {A} / {B} is {result}")
    except ZeroDivisionError:
        print("infinite")

# Example usage
A = int(input("Enter the value of A: "))
B = int(input("Enter the value of B: "))
divide(A, B)
```
